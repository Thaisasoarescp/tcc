```python
!pip install feedparser
import feedparser
import pandas as pd
import time
from urllib.parse import quote
from datetime import datetime, timedelta

def coletar_noticias_excel(sites, anos=5):
    hoje = datetime.now()
    # Ponto de partida: 5 anos atrás
    data_inicio = hoje - timedelta(days=anos*365)

    # Criando intervalos mensais para "driblar" o limite do Google
    intervalos = pd.date_range(start=data_inicio, end=hoje, freq='MS')

    lista_final = []

    print(f"🚀 Iniciando coleta histórica de {anos} anos...")

    for i in range(len(intervalos)-1):
        inicio = intervalos[i].strftime('%Y-%m-%d')
        fim = intervalos[i+1].strftime('%Y-%m-%d')

        # Query refinada: Clima + Portais específicos
        query = f'("mudanças climáticas" OR "aquecimento global") (site:uol.com.br OR site:jovempan.com.br) after:{inicio} before:{fim}'

        url = f"https://news.google.com/rss/search?q={quote(query)}&hl=pt-BR&gl=BR&ceid=BR:pt-419"

        feed = feedparser.parse(url)

        for entry in feed.entries:
            lista_final.append({
                "Portal": entry.source.title if 'source' in entry else "N/A",
                "Título": entry.title.split(" - ")[0],
                "URL": entry.link,
                "Data": entry.published
            })

        print(f"📅 Período {inicio} a {fim} concluído. ({len(lista_final)} itens acumulados)")
        time.sleep(1) # Pausa para evitar bloqueio

    # Criando o DataFrame
    df = pd.DataFrame(lista_final)

    if not df.empty:
        # Tratamento de dados: Removendo duplicatas e formatando data
        df = df.drop_duplicates(subset=['URL'])
        df['Data'] = pd.to_datetime(df['Data'], errors='coerce').dt.strftime('%d/%m/%Y')

        # Salvando em Excel
        nome_arquivo = "noticias_clima_uol_jp_5anos.xlsx"
        df.to_excel(nome_arquivo, index=False, engine='openpyxl')

        print(f"\n✅ Sucesso! Arquivo '{nome_arquivo}' gerado com {len(df)} linhas.")
    else:
        print("\n❌ Nenhuma notícia encontrada no período.")

# --- Execução ---
portais = ["uol.com.br", "jovempan.com.br"]
coletar_noticias_excel(portais)
```
